```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.biztory.com/about'
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
print(soup)
```

```python
rows = soup.find_all("div", class_="m-member__overlay p-5 p-md-15")
```

```python
rows
```

```python
rows2 = soup.find_all("p", class_="mb-5")
```

```python
rows2[:5]
```

```python
rows5 = []
for i in rows2:
    rows5.append(i.split("<p")[1])
rows5
```

```python
rows3 = soup.find_all("strong")
```

```python
rows4 = rows3[2:-7]
rows4
```

```python
rows6 = []
for i in rows4:
    rows6.append(i.split("<strong>")[1])
rows6
```
